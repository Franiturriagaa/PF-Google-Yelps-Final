```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('../../Data/business.pkl')
```

```python
df=df.loc[:, ~df.columns.duplicated()]
```

```python
df
```

```python
df = df[(df['city'] =='Philadelphia')]
```

```python
df
```

```python
df.drop(columns={'is_open', 'attributes', 'hours'}, inplace=True)
```

```python
df['categories'] = df['categories'].str.split(',\s*')
```

```python
df.dropna(subset=['categories'], inplace=True)
```

```python
main_categories_set = {'Restaurants', 'Shopping', 'Bars', 'Beauty & Spas', 'Home Service', 'Health & Medical','Food','Home Services','Automotive', 'Local Services', 'Active Life', 'Hotels & Travel', 'Pets', 'Arts & Entertainment'}

def get_main_category(categories):
    for category in categories:
        if category in main_categories_set:
            categories.remove(category)  
            return category, categories  
    return None, categories
```

```python
df[['main category', 'categories']] = df.apply(lambda row: pd.Series(get_main_category(row['categories'])), axis=1)
```

```python
df
```

```python
df['main category'].isnull().sum()
```

```python
df_test = df[(df['main category'].isnull())]
```

```python
df_exploded = df_test.explode('categories')
```

```python
category_counts = df_exploded['categories'].value_counts()
```

```python
top_categories   = category_counts.head(20)
```

```python
plt.figure(figsize=(14, 8))

top_categories.plot(kind='bar', color='skyblue')

plt.xlabel('Categoria')
plt.ylabel('Cantidad')
plt.title('Top 20 Categorias mas comunes')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
